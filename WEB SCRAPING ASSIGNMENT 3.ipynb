### Que.3- In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
#extract the tag having urls of all pages
url = web_driver.find_elements_by_xpath("//ul[@class='a-pagination']//a")

sunglasses_urls = [] #empty list

#to remove extra data other than output we required
for i in url[0:3]:
    sunglasses_urls.append(i.get_attribute('href'))
sunglasses_urls
```

```python
Brand_name = [] #empty list
#extract the tag having brand name
for i in sunglasses_urls:
    web_driver.get(i)
    brand = web_driver.find_elements_by_xpath("//span[@class='a-size-base-plus a-color-base']")
    for i in brand:
        Brand_name.append(i.text) #to remove extra data other than output we required
```

```python
len(Product_name)
```

```python
Product_name = [] #empty list
#extract the tag having brand name
for i in sunglasses_urls:
    web_driver.get(i)
    brand = web_driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']")
    for i in brand:
        Product_name.append(i.text) #to remove extra data other than output we required
```

```python
#extract the tag having urls of all compaies
url = web_driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")

rating_urls = [] #empty list
#to remove extra data other than output we required
for i in sunglasses_urls:
    for j in url:
        rating_urls.append(j.get_attribute('href'))
```

```python
Rating = [] #empty list
#extract the tag having brand name
for i in rating_urls:
    web_driver.get(i)
    rate = web_driver.find_elements_by_xpath("//span[@class='a-size-base a-nowrap']")
    for i in rate:
        Rating.append(i.text.replace(' ', '-')) #to remove extra data other than output we required
Rating
```

```python
Num_rating = [] #empty list
#extract the tag having brand name
for i in sunglasses_urls:
    web_driver.get(i)
    N_R = web_driver.find_elements_by_xpath("//span[@class='a-size-base']")
    for i in N_R:
        Num_rating.append(i.text) #to remove extra data other than output we required
```

```python
Price = [] #empty list
#extract the tag having brand name
for i in sunglasses_urls:
    web_driver.get(i)
    N_R = web_driver.find_elements_by_xpath("//span[@class='a-price-whole']")
    for i in N_R:
        Price.append(i.text) #to remove extra data other than output we required
```

```python
Return = [] #empty list
#extract the tag having brand name
for i in rating_urls:
    web_driver.get(i)
    retur = web_driver.find_elements_by_xpath("//div[@class='a-section a-spacing-none icon-content']")
    for i in retur:
        Return.append(i.text.replace(' ', '-')) #to remove extra data other than output we required
Return
```

```python
Availability = [] #empty list
#extract the tag having brand name
for i in rating_urls:
    web_driver.get(i)
    retur = web_driver.find_elements_by_xpath("//div[@class='a-section a-spacing-base }']")
    for i in retur:
        Availability.append(i.text.replace(' ', '-')) #to remove extra data other than output we required
Availability
```

```python
other_details = [] #empty list
#extract the tag having brand name
for i in sunglasses_urls:
    web_driver.get(i)
    details = web_driver.find_elements_by_xpath("//div[@class='a-row a-size-base a-color-secondary s-align-children-center']")
    for i in details:
        other_details.append(i.text) #to remove extra data other than output we required
other_details
```

```python
#extract the tag having urls of all compaies
url = web_driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")

Product_url = [] #empty list
#to remove extra data other than output we required
for i in sunglasses_urls:
    for j in url:
        Product_url.append(j.get_attribute('href'))
Product_url
```

```python
print(len(Brand_name), len(Product_name), len(Rating), len(Num_rating), len(Price), len(Return), len(Availability), len(other_details), len(Product_url))
```

```python
# Make data frame of sunglasses from Amazon.com
Sunglasses = pd.DataFrame({})
Sunglasses['Brand Name']= Brand_name[0:144]
Sunglasses['Product Name'] = Product_name[0:144]
Sunglasses['Rating'] = Rating[0:144] 
Sunglasses['Number of Ratings'] = Num_rating[0:144]
Sunglasses['Price'] = Price[0:144] 
Sunglasses['Return'] = Return[0:144] 
Sunglasses['Availability'] = Availability[0:144]
Sunglasses['Saving Details'] = other_details[0:144] 
Sunglasses['Product URL'] = Product_url 
Sunglasses
```

```python
Sunglasses.to_csv("sunglasses.csv")
```

### Que.4- Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page.
